```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

## HW 4

### Problem 0 - Setting up data

```python
t = np.linspace(0, np.pi*100, np.int(1e5))
```

```python
N = 10 + np.sin(t)
```

```python
bkgd = stats.norm.rvs(size=np.int(1e5)) * np.sqrt(N) + N
```

## Problem 1 - Subsets of data

```python
first_1000 = bkgd[0:1000]
plt.scatter(t[0:1000], first_1000)
plt.ylabel('Background Data')
plt.xlabel('Time')
plt.title('Scatter plot of first 1000 data points')
```

```python
def plot_n(start, end):
    first_n = bkgd[start:end]
    time = t[start:end]
    plt.scatter(time, first_n)
    plt.xlabel('Time')
    plt.ylabel('Background Data')
    plt.title('Scatter plot of background data from % 2d to % 3d' %(start, end))
```

```python
plot_n(49500,50500)
```

```python
data = []
time = []
for i in range(0, 101):
    data = np.append(data, bkgd[100 * i])
    time = np.append(time, t[100 * i])
```

```python
plt.scatter(time, data)
plt.xlabel('Time')
plt.ylabel('Background Data')
plt.title('Scatter plot of every 100th point')
```

### Problem 2 - Density

```python
_ = plt.hist2d(t, bkgd, bins = 25, density = True)
plt.colorbar()
plt.ylabel('Background Data Density')
plt.xlabel('Time')
plt.title('Density of background data')
```

We are plotting the number of points in a region relative to the overall data in the background. Each color shows the specific density of the region and each color is specific to a certain density. THe brighter the color the higher the density which means that there are more points concentrated in that region in comparison to others.

### Problem 3 - Folding

```python
new_t = t % (2*np.pi)
plt.scatter(new_t, bkgd)
plt.xlabel('Time')
plt.ylabel('Background Data')
plt.title('Scatter plot of data over folded time')
```

```python
_ = plt.hist2d(new_t, bkgd, bins = 25, density = True)
plt.colorbar()
plt.xlabel('Time')
plt.ylabel('Background Data Density')
plt.title('Scatter plot of data over folded time')
```

We can see that the background data now resembles a more periodic signal.
